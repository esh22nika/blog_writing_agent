## Imports

````python
from __future__ import annotations

import json
import operator
from typing import TypedDict, List, Annotated
from pathlib import Path

from pydantic import BaseModel, Field

from langgraph.graph import StateGraph, START, END
from langgraph.types import Send

from langchain_ollama import ChatOllama  # Modern Ollama integration
from langchain_core.messages import HumanMessage
````

## LLM Setup

Using `ChatOllama` with `qwen2.5`. Change the model name here if you want to use a different one (e.g. `llama3.2`, `mistral`).

````python
llm = ChatOllama(
    model="qwen2.5",   # change to any model you have pulled in ollama
    temperature=0.3
)

# Quick sanity check
print(llm.invoke("Say hello in one sentence.").content)
````

````python
research_llm = ChatOllama(
    model="qwen2.5",
    temperature=0
)

writer_llm = ChatOllama(
    model="qwen2.5",
    temperature=0.3
)
````

## Data Models

````python
class Task(BaseModel):
    id: int
    title: str

    goal: str = Field(
        ...,
        description="What the reader should understand after this section."
    )

    bullets: List[str] = Field(
        ...,
        min_length=3,
        max_length=5,
        description="Concrete subpoints the section must cover."
    )

    target_words: int = Field(
        ...,
        description="Approximate length of the section."
    )


class Plan(BaseModel):
    blog_title: str
    audience: str = Field(..., description="Who the blog is for")
    tone: str = Field(..., description="Writing tone")
    tasks: List[Task]
````

## LangGraph State

````python
class State(TypedDict):
    topic: str
    plan: Plan
    research: str
    sections: Annotated[List[str], operator.add]
    images: Annotated[List[str], operator.add]
````

## Agent Nodes

### 1. Orchestrator creates the blog plan

````python
def orchestrator(state: State) -> dict:

    prompt = f"""
You are a senior technical writer.

Create a detailed blog plan for the topic: {state['topic']}

Requirements:
- 5–7 sections
- Each section must include:
  1. title
  2. goal (what reader learns)
  3. 3–5 bullet points
  4. target word count (120–450)

Return ONLY valid JSON with this format:

{{
  "blog_title": "...",
  "audience": "...",
  "tone": "...",
  "tasks": [
    {{
      "id": 1,
      "title": "...",
      "goal": "...",
      "bullets": ["...", "..."],
      "target_words": 200
    }}
  ]
}}
"""

    response = research_llm.invoke(prompt).content.strip()

    # Remove markdown fences if model adds them
    response = response.replace("```json", "").replace("```", "").strip()

    # Extract JSON block if extra text exists
    start = response.find("{")
    end = response.rfind("}") + 1
    response = response[start:end]

    plan_dict = json.loads(response)

    
    plan = Plan(**plan_dict)
    print(f"[Orchestrator] Plan created: '{plan.blog_title}' with {len(plan.tasks)} sections")
    return {"plan": plan}
````

### 2. Fan-out (sends each task to a worker in parallel)

````python
def fanout(state: State):
    """Distributes tasks across parallel worker nodes."""
    return [
        Send(
            "worker",
            {
                "task": task,
                "topic": state["topic"],
                "plan": state["plan"],
                "research": state["research"]
            }
        )
        for task in state["plan"].tasks
    ]
````

````python
def researcher(state: State) -> dict:

    topic = state["topic"]

    prompt = f"""
You are a research assistant.

Gather key technical facts and insights for writing a blog about:

{topic}

Return concise research notes including:
- important concepts
- definitions
- relevant examples
- key terminology
- modern developments

Write in bullet points.
"""

    research = research_llm.invoke(prompt).content.strip()

    print("[Research] completed")

    return {"research": research}
````

````python
def image_generator(state: State) -> dict:

    prompt = f"""
Generate 3 image search queries for a blog about:

{state["topic"]}

Return ONLY a JSON list.

Example:
[
 "self attention transformer diagram",
 "transformer architecture visualization",
 "attention mechanism illustration"
]
"""

    queries = research_llm.invoke(prompt).content

    try:
        queries = json.loads(queries)
    except:
        queries = []

    images = []

    for q in queries:
        url = f"https://source.unsplash.com/featured/?{q.replace(' ', ',')}"
        images.append(f"![{q}]({url})")

    print("[Images] generated")

    return {"images": images}
````

### 3. Worker writes one blog section

````python
def worker(payload: dict) -> dict:

    task = payload["task"]
    topic = payload["topic"]
    plan = payload["plan"]
    research = payload["research"]

    bullets_text = "\n- " + "\n- ".join(task.bullets)

    prompt = f"""
You are writing ONE section of a technical blog.

Blog title: {plan.blog_title}
Audience: {plan.audience}
Tone: {plan.tone}

Use this research when writing:
{research}

Section title: {task.title}

Goal:
{task.goal}

Cover these bullets:
{bullets_text}

Target length: {task.target_words} words

Return clean markdown starting with:

## {task.title}
"""

    section_md = writer_llm.invoke(prompt).content.strip()

    print(f"[Worker] Done: {task.title}")

    return {"sections": [section_md]}
````

### 4. Reducer stitches everything into the final blog

````python
def reducer(state: State):

    blog = "# " + state["plan"].blog_title + "\n\n"

    if state["images"]:
        blog += state["images"][0] + "\n\n"

    for i, section in enumerate(state["sections"]):

        blog += section + "\n\n"

        if i + 1 < len(state["images"]):
            blog += state["images"][i + 1] + "\n\n"

    filename = state["plan"].blog_title.lower().replace(" ", "_") + ".md"

    with open(filename, "w", encoding="utf-8") as f:
        f.write(blog)

    print(f"Saved {filename}")

    return {}
````

## Build & Compile the Graph

````python
g = StateGraph(State)
g.add_node("researcher", researcher)
g.add_node("orchestrator", orchestrator)
g.add_edge(START, "researcher")
g.add_edge("researcher", "orchestrator")
g.add_node("worker", worker)
g.add_node("reducer", reducer)
g.add_node("image_generator", image_generator)

g.add_edge(START, "orchestrator")
g.add_conditional_edges("orchestrator", fanout, ["worker"])
g.add_edge("worker", "reducer")
g.add_edge("image_generator", "reducer")

g.add_edge("reducer", END)

app = g.compile()
app
````

## Run the Agent

````python
out = app.invoke({"topic": "Write a blog on Self Attention", "sections": []})
````
